```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('AirQualityUCI.csv')
df
```

```python
df
```

```python
df1=df.loc[0:25]
```

```python
df1
```

```python
plt.figure(figsize=(15,9))
sns.boxplot(data=df1,x=df1['Time'],y=df1['T'])
plt.xlabel("COgt")
plt.ylabel("T")
plt.title("Co vs T")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.histplot(df['NOx(GT)'],kde=True,bins=10)
plt.xlabel("COgt")
plt.ylabel("T")
plt.title("Co vs T")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x=df['RH'],y=df['AH'])
plt.xlabel("COgt")
plt.ylabel("T")
plt.title("Co vs T")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df1,x=df1['Time'],y=df1['T'])
plt.xlabel("COgt")
plt.ylabel("T")
plt.title("Co vs T")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df1,x=df1['T'],y=df1['AH'],hue='C6H6(GT)')
plt.xlabel("COgt")
plt.ylabel("T")
plt.title("Co vs T")
plt.show()
```

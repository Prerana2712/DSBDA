```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
df
```

```python
df1=df.loc[0:50]
```

```python
plt.figure(figsize=(15,9))
sns.boxplot(data=df1,x=df1['petal_length'],y=df1['petal_width'])
plt.xlabel("Petal Length")
plt.ylabel("petal width")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.histplot(data=df,x=df['sepal_length'],bins=10,kde=True)
plt.xlabel("sepal Length")
plt.ylabel("Freq")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(df['sepal_width'],label
             ='Sepal Width')
plt.xlabel("sepal Length")
plt.ylabel("Freq")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x=df['sepal_width'],y=df['sepal_length'])
plt.xlabel("sepal Length")
plt.ylabel("Freq")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(df['sepal_width'],label='Sepal Width')
sns.lineplot(df['sepal_length'],label='Sepal lengt')
sns.lineplot(df['petal_length'],label='Petal length')
sns.lineplot(df['petal_width'],label='Petal Width')
plt.xlabel("sepal Length")
plt.ylabel("Freq")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x=df1['sepal_width'],y=df1['sepal_length'],hue='species')
plt.xlabel("sepal Length")
plt.ylabel("Freq")
plt.show()
```

```python
sns.countplot(x=df['species'])
```

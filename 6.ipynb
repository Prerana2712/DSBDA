```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
df
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=df.target,y=df.age,data=df)
plt.xlabel('target')
plt.ylabel('age')
plt.title('Age vs target')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df['chol'],bins=10,kde=True)
plt.xlabel('Cholestrol')
plt.ylabel('Frequency')
plt.title('histo of chol')
plt.show()
```

```python
plt.figure(figsize=(15,9))
plt.plot(df['age'])
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Resting BP vs Age')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x=df['age'],y=df['fbs'])
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Resting BP vs Age')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x=df['age'],y=df['chol'],hue='target')
plt.xlabel('Age')
plt.ylabel('chol')
plt.title('Resting BP vs Age')
plt.show()
```

```python
sns.lineplot(data=df[['chol','thal']])
```

```python
plt.figure(figsize=(20,15))
pd.crosstab(df.age,df.target).plot(kind='bar')
plt.xlabel('age')
plt.ylabel('cholestrol')
plt.title('Age vs cholestrol')
plt.show()
```

```python
sns.countplot(data=df,x=df['target'])
```
